# Exercise 2

- With faang, use type conversion to change the date column into a datetime and the volume column into integers. Then, sort by date and ticker.
- Find the seven rows with the highest value for volume.
- Right now, the data is somewhere between long and wide format. Use melt() to make it completely long format. Hint: date and ticker are our ID variables (they uniquely identify each row). We need to melt the rest so that we don't have separate columns for open, high, low, close, and volume.

```python
# With faang, use type conversion to change the date column into a datetime and
# the volume column into integers. Then, sort by date and ticker.
faang = p.read_csv('/content/faang.csv')
new_faang = faang.assign(
    date=p.to_datetime(faang.date),
    volume = faang['volume'].astype('int')
)
sorted_new_faang = new_faang.sort_values(by=['date', 'Ticker'])
sorted_new_faang
```

```python
# Find the seven rows with the highest value for volume.
faang_max = faang.nlargest(n=7, columns='volume')
faang_max
```

```python
# Right now, the data is somewhere between long and wide format. Use melt() to make
# it completely long format. Hint: date and ticker are our ID variables (they uniquely identify each row).
# We need to melt the rest so that we don't have separate columns for open, high, low, close, and volume
melted_faang = new_faang.melt(
    id_vars = ['date', 'Ticker']
)
melted_faang
```
