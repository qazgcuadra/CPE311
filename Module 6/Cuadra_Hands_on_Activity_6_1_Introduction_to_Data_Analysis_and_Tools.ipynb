```python
# Median funciton
def median(data):
    data.sort() # sorts the dataset
    if len(data) % 2 != 0: # checks if the median is Odd
        median = data[len(data)//2] # finding the index of the list by diving the length with 2
    else: # if the median is even
        median = data[(int(len(data)/2)) - 1] + data[int(len(data)//2)] # it adds the median with the index before it
        median = median//2
    return median

median(salaries)
```

```python
def mode(data):
    counter = 0 # keeps track of the maximum frequency in the list
    num = data[0] # this stores the most frequent number on the list

    for i in data:
        n = data.count(i) # to count the number of times an element appears
        if n > counter: # checks if the current element is greater than the maximum frequency
            counter = n
            num = i # updates the maximum frequency to the current element
        if len(set(data)) == len(data): # set function does not allow repeating elements
            return 'Mode is non-existent' # if the set data is equivalent to the length of the dataset then it returns the following
    return num

mode(salaries)
```

```python
import statistics as st

st.mode(salaries)
```

```python
# Sample Variance function
def s_var(data):
    mean = sum(data) / len(data)
    value = 0 # used to store the squared differences
    for i in data:
        value += (i - mean)**2 # the squared difference is then added to the result current squared difference which then updates the value
    s_var = value / (len(data) - 1) # calculation for sample variance
    return s_var

s_var(salaries)
```

```python
import statistics as st

st.variance(salaries)
```

```python
# Sample standard deviation function
def SSD(data):
    mean = sum(data) / len(data)
    value = 0
    for i in data:
        value += (i - mean)**2
    s_var = (value / (len(data) - 1))**0.5 # the calculation for sample variance is then raised to 0.5 to square it
    return s_var

SSD(salaries)
```

```python
import statistics as st

st.stdev(salaries)
```

Using the same data, calculate the following statistics using the functions in the statistics module where appropriate:

- Range
- Coefficient of variation
- Interquartile range
- Quartile coefficient of dispersion

```python
# Range function
def range(data):
    return max(data) - min(data) # returns the difference between the maximum and minimum data

range(salaries)
```

```python
# Coefficient of variation
import statistics as st

def coeff_var(data):
    st_dev = st.stdev(data)
    mean = st.mean(data)
    return (st_dev/mean) * 100 # returns the calculation for the coefficient of variation

coeff_var(salaries)
```

```python
# Interquartile Range function
import statistics as stats

def IQR(data):
    IQR = st.quantiles(data, n=4)[-1] - st.quantiles(data, n = 4)[0] # gets the difference of the upper and lower middle elements. 4 is used for quartile
    return IQR # returns the calculation for the Interquartile Range


IQR(salaries)
```

```python
# Quartile coefficient of dispersion
import statistics as stats

def QCD(data):
    qrt = st.quantiles(data, n=4)
    iqr = qrt[-1] - qrt[0]
    return iqr/(qrt[-1] + qrt[0]) # returns the calculated QCD

def QCD_percent(data):
    qrt = st.quantiles(data, n=4)
    iqr = qrt[-1] - qrt[0]
    return iqr/(qrt[-1] + qrt[0]) * 100 # # returns the calculated QCD percentage

print(QCD(salaries), QCD_percent(salaries))
```
