```python
# Renaming columns
df.rename(columns={'Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)' :
                   'Total Alcohol Consumptions', 'Entity' : 'Country'}, inplace=True)

df
```

```python
# Creation of an additional column that outputs 'Yes' if the year is equal or above 2010 and 'No' if not
df['2010 and above'] = np.where(df['Year'] >= 2010, 'Yes', 'No')
df
```

```python
# Displaying the values above the year or equal 2010
above2010 = df[df['2010 and above'] == 'Yes']

above2010
```

```python
# Displaying the values below the year 2010
below2010 = df[df['2010 and above'] == 'No']

below2010
```

```python
# Displaying the values between 2010 and 2015
between2010_2015 = df[df['Year'].between(2010, 2015)]

between2010_2015
```

```python
# Average year
np.mean(df['Year'])
```

```python
# Year Median
np.median(df['Year'])
```

```python
# Year Standard Deviation
np.std(df['Year'])
```

```python
import statistics as st

def IQR(data):
    IQR3 = st.quantiles(data, n=4)[-1]
    print(IQR3)
    IQR1 = st.quantiles(data, n = 4)[0]
    print(IQR1) # gets the difference of the upper and lower middle elements. 4 is used for quartile
    return IQR3 - IQR1 # returns the calculation for the Interquartile Range


IQR(df['Year'])
```
