2. Load the dataset into pandas dataframe.

```python
import pandas as pd
import numpy as np

co2e = pd.read_csv(
    '/content/Total GHG by Sector.csv',
)
co2e
```

3. Prepare the data by applying appropriate data preprocessing techniques.

```python
# filtering out the datas that are lower than 2001
co2e2000s = co2e.query('Year > 2000')
co2e2000s
```

```python
# transforming Year column into index
co2e2000s.set_index('Year', inplace=True)
co2e2000s
```

```python
# checking if the Year column turned into index
co2e2000s.dtypes
```

```python
# adds a column "Total" that adds the sum of the rows
co2e2000s['Total'] = co2e2000s.sum(axis=1)
co2e2000s
```

```python
# sort the dataframe by descending order of the total column
sort_co2 = co2e2000s.sort_values(by='Total', ascending=False)
sort_co2
```

```python
# displays the top 10 years that produced the most CO2
sort_head_co2 = sort_co2.head(10)
sort_head_co2
```

4. Analyze the data using descriptive analysis.

```python
# displays the descriptive data analysis of the top 10 years and its total
sort_head_co2.describe()
```

5. Perform correlation analysis.

```python
sort_head_co2.corr()
```

```python
# displaying the correlation in a plot
from pandas.plotting import lag_plot

lag_plot(sort_head_co2)
```

6. Interpret the results based on the descriptive and correlation analysis.

- It displays a strong positive correlation of the greenhouse emissions. It also shows the averages of different greenhouse emissions per year as well as the fluctuation of values in the years were CO2 emissions are in their all time high. The standard deviation provides insights on the amount of emissions that deviate from their respective averages during its peak. This benefits us in a way that we can identify the consistency of greenhouse emission during these pivotal years.
